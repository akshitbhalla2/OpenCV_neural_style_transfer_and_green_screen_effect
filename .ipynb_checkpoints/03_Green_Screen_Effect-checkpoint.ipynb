### Import Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

### Read - Display Image

```python
# Read the image
img = cv2.imread("./images/car.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img.shape
```

```python
# Make a copy of the image
img_new = img.copy()
```

## Define Color Threshold

```python
# Isolate the Green Background
# Try it by yourselves. Change the values to getyour best result!
lower_green = np.array([0, 200, 0])
upper_green = np.array([250, 255, 250])
```

## Creat the Mask

```python
# Define the Masked Area
mask = cv2.inRange(
    img_new,
    lower_green,
    upper_green
)

# Visualize the Mask on the imgae
plt.imshow(mask)
```

```python
# Mask the Image to let your object come and show through
img_mask = img_new.copy()
img_mask[mask != 0] = [0, 0, 0]

# Dispaly the Image
plt.imshow(img_mask)
```

## Mask and Add a Background Image

```python
# Load the Image Background
img_back = cv2.imread("./images/london.jpg")

# Convert it from BGR to RGB
img_back = cv2.cvtColor(img_back, cv2.COLOR_BGR2RGB)

# Crop it to the the dimensions we got
img_back = img_back[0:img.shape[0], 0:img.shape[1]]

# Mask the Cropped Background
# That's how the object's area is blocked
img_back[mask == 0] = [0, 0, 0] 

# Display the Background
plt.imshow(img_back)
```

## Let's move it

```python
# x, y, h, w
x, y, h, w = 390, 300, 450, 660

# Read and convert it from BGR to RGB
img_crop = cv2.imread("./images/london.jpg")
img_crop = cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB)

# Crop it to the the dimensions we got
img_crop = img_crop[y:y+h, x:x+w]

# Mask the Cropped Background
# That's how the object's area is blocked
img_crop[mask == 0] = [0, 0, 0] 

# Display the Background
plt.imshow(img_crop)
```

# Add them together and get the Final Image

```python
# Add 
img_complete = img_back + img_crop
plt.imshow(img_complete)
```
